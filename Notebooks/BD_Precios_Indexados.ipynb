## Cleaning Data

#### Managing NaN

First we substitue the "Precio", "Fecha_Finalización" and "Comercializadora" NaNs for value 0. 
The idea is to look for the value 0 in the column of dates when we look for a date to perform the analisis. 

```python
precio_indexado_df.replace(np.nan, 0, inplace=True)
```

```python
precio_indexado_df.isna().sum()
```

or, we live the NaN values as it will show when a factura de luz has not been inserted properly and when looking for "Fecha_Finalización" and "Comercializadora" we can search for value NaN instead of 0.  

#### Changing "Date" column from object to date type:

```python
precio_indexado_df['Fecha_Inicio'] = pd.to_datetime(precio_indexado_df['Fecha_Inicio'], infer_datetime_format=True)
precio_indexado_df['Fecha_Inicio']
```

```python
precio_indexado_df["Fecha_Finalización"] = np.where((precio_indexado_df["Fecha_Finalización"] == "30.02.2018"), "28.02.2018",precio_indexado_df["Fecha_Finalización"]) 
precio_indexado_df["Fecha_Finalización"] = np.where((precio_indexado_df["Fecha_Finalización"] == "30.02.2019"), "28.02.2019",precio_indexado_df["Fecha_Finalización"]) 
precio_indexado_df["Fecha_Finalización"] = np.where((precio_indexado_df["Fecha_Finalización"] == "30.02.2020"), "28.02.2020",precio_indexado_df["Fecha_Finalización"]) 
precio_indexado_df["Fecha_Finalización"]
```

```python
precio_indexado_df["Fecha_Finalización"].unique()
```

```python
precio_indexado_df['Fecha_Finalización'] = pd.to_datetime(precio_indexado_df['Fecha_Finalización'], infer_datetime_format=True)
precio_indexado_df['Fecha_Finalización']
```

#### Changing all "Precio_Potencia" columns values  that are  MWh/€ to KWh/€  and doing the same for Precio_Energía to check. 

All price values higher than 0 we divide by 1000 to transform to KWh. 

```python
precio_indexado_df['Precio_Potencia_P1'] = precio_indexado_df['Precio_Potencia_P1'].apply(lambda x: x/1000 if x>0.5 else x)
precio_indexado_df['Precio_Potencia_P2'] = precio_indexado_df['Precio_Potencia_P2'].apply(lambda x: x/1000 if x>0.5 else x)
precio_indexado_df['Precio_Potencia_P3'] = precio_indexado_df['Precio_Potencia_P3'].apply(lambda x: x/1000 if x>0.5 else x)
precio_indexado_df['Precio_Potencia_P4'] = precio_indexado_df['Precio_Potencia_P4'].apply(lambda x: x/1000 if x>0.5 else x)
precio_indexado_df['Precio_Potencia_P5'] = precio_indexado_df['Precio_Potencia_P5'].apply(lambda x: x/1000 if x>0.5 else x)
precio_indexado_df['Precio_Potencia_P6'] = precio_indexado_df['Precio_Potencia_P6'].apply(lambda x: x/1000 if x>0.5 else x)
precio_indexado_df['Precio_Potencia_P6'] = precio_indexado_df['Precio_Potencia_P6'].apply(lambda x: x/1000 if x>0.5 else x)
precio_indexado_df['Precio_Energía_P1'] = precio_indexado_df['Precio_Energía_P1'].apply(lambda x: x/1000 if x>0.5 else x)
precio_indexado_df['Precio_Energía_P2'] = precio_indexado_df['Precio_Energía_P1'].apply(lambda x: x/1000 if x>0.5 else x)
precio_indexado_df['Precio_Energía_P3'] = precio_indexado_df['Precio_Energía_P1'].apply(lambda x: x/1000 if x>0.5 else x)
precio_indexado_df['Precio_Energía_P4'] = precio_indexado_df['Precio_Energía_P1'].apply(lambda x: x/1000 if x>0.5 else x)
precio_indexado_df['Precio_Energía_P5'] = precio_indexado_df['Precio_Energía_P1'].apply(lambda x: x/1000 if x>0.5 else x)
precio_indexado_df['Precio_Energía_P6'] = precio_indexado_df['Precio_Energía_P1'].apply(lambda x: x/1000 if x>0.5 else x)
```
